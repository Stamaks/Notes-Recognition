```python
import numpy as np
from librosa import load, logamplitude
from librosa.feature import melspectrogram
from tqdm import tqdm_notebook
```

```python
def get_spectrogram(path):
    """Строим спектограмму из wav файла"""
    y, sr = load(path)
    S = melspectrogram(y, sr=sr, n_mels=100)
    log_S = logamplitude(S, ref_power=np.max)
    return log_S
```

```python
d = {}

for i in range (len(spectrograms)):
#   Режем каждую спектрограмму на кусочки по 8 (примерно 0.2 секунды)
    name = files_names[i][2:-12]
    dict.update(d, {name : [spectrograms[i][:, a:a+8] for a in range(len(spectrograms[i][0]) // 8)]})

len(list(dict.items(d))[0][1])
```

### Потренируем модели на маленьких кусочках

```python
number_of_notes = 88
number_of_sets = 5
data_answ = [((len(spectrograms), number_of_notes))]
```

```python
data_answ = [[0 for a in range(number_of_notes)] for b in range(len(spectrograms))]
```

```python
len(data_answ)
```

```python
items = list(dict.items(d))
items[0][0]
```

```python
for i in range (len(items)):
#   Берем инфорацию из имени файла, записываем в матрицу единицы напротив звучащих в файле нот
#   file name example 2,23,36,120.mid.wav
    name = items[i][0].split(',')
    for j in range (len(name)):
        if (int(name[j]) >= 21 and int(name[j]) < 110):
            data_answ[i][int(name[j]) - 21] = 1
```

```python
data_answ = data_answ*number_of_sets
```

```python
data_answ = np.array(data_answ, dtype=int)
data_answ.shape
```

```python
import random
a = []
for i in range(number_of_sets):
    for j in range(len(items)):
        b = random.randint(0, len(items[j][1]) - 1)
        c = items[j][1][b]
        c = c.ravel()
        a.append(c)
```

```python
a = np.array(a)
a.shape
```

```python
random_number = len(spectrograms)*3
X_train = a[:random_number]
Y_train = data_answ[:random_number]
X_test = a[random_number:]
Y_test = data_answ[random_number:]
```

```python
from sklearn.linear_model import LogisticRegression
from tqdm import tqdm_notebook
```

```python
models = []
if os.stat("Models_trained_piano_wn_little_2").st_size != 0:
    models = pickle.load(open("Models_trained_piano_wn_little_2", 'rb'))
```

```python
if len(models) != 88 :
    models = []
    for i in tqdm_notebook(range(number_of_notes)):
        model = LogisticRegression()
        model.fit(X_train, Y_train[:, i])
        models.append(model)
    pickle.dump(models, open("Models_trained_piano_wn_little_2", "wb"))
```

```python
Y_pred_trains = []
Y_pred_tests = []
for i in tqdm_notebook(range(88)):
    Y_pred_tests.append(models[i].predict(X_test))
    Y_pred_trains.append(models[i].predict(X_train))
```

```python
#Процент полностью отгаданных правильно аккордов
#Вчера был 0.41, лол. Все из-за рандомного датасета .-.
np.mean((np.array(Y_pred_trains).T* Y_train).sum(1) / Y_train.sum(1))
```

```python
from sklearn.metrics import roc_auc_score
train = []
test = []
for i in tqdm_notebook(range(88)):
    train.append(roc_auc_score(Y_train[:, i],Y_pred_trains[i]))
    test.append(roc_auc_score(Y_test[:, i],Y_pred_tests[i]))
print ("Среднее на обучении", np.mean(train))
print ("Среднее на    тесте", np.mean(test))
```
